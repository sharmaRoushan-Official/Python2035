## 3. Lists in Python

--

* A `list` in Python is a mutable, ordered, indexed collections that can store multiple values(inclduding mixed data types). 
* Lists are one of the most used data types in Python.

## 1. List Definition


```python
my_list = [1,2,3,4]
print(my_list)
print(type(my_list))
```

```python
empList = []
print(empList)
print(type(empList))
```

```python
data = [10,20,30,40]
print(data)
print(type(data))
```

### 2. Features of List

---

|Feature|Explanation|
|---|---|
|Ordered|Items maintain insertion order|
|Indexed|Each elements has an index|
|Mutable|Can be modified after creation|
|Allows duplicates|Same values can repeat|
|Mixed types allowed|int,float,str,bool, list, tuple, set, dictionary etc.

```python
mix = [10,25.5,"Python",True]
print(mix)
```

## 3. List Indexing

1. Positive Indexing
2. Negative Indexing

```python
nums = [100,200,300,400]
print(nums)
```

```python
## Negative Indexing
print(nums[-1])
print(nums[-2])
print(nums[-3])
print(nums[-4])
# print(nums[-5])
```

### 6. Adding Elements to a List

A). append(): Adds element at the end one by one.

B). insert(): Adds element at specific index

C). extend(): Adds multiple elements

### A). append(): Adds element at the end one by one.


```python
nums = [1,2,3]
nums.append(4)
print(nums)
```

```python
nums.append(4,5)
print(nums)
```

### B). insert(): Adds element at specific index

```python
nums = [10,20,30]
print(nums)
print(type(nums))
```

```python
nums.insert(0,-20)
print(nums)
```

### C).extend() --> Adds multiple elements


```python
a = [1,2,3]
b = [4,5,6] 
a.extend(b)  # merging two lists
print(a)
```

### 7. Removing Elements from list:

A). remove(): Remove by value:

B). pop(n): Remove by index value(returned removed element)

C). pop(): without index --> Removes last elements

D). clear(): Removes all elements at once.

E). del() : Delete an items by index value.

### A). remove(): Remove by value:

```python
nums = [10,20,30,20]
nums.remove(20)  # Removes only the first occurrence.
print(nums)
```

### B). pop(n): Remove by index value(returned removed element)

```python
nums = [10,20,30,40,20]

nums.pop(4)  # 4- index value
```

```python
print(nums)
```

### C). pop(): without index --> Removes last elements


```python
nums = [10,20,30,40]
print(nums)
```

```python
nums.pop()
```

```python
print(nums)
```

### D). clear(): Removes all elements at once.

```python
nums = [10,20,30,40]
print(nums)
```

```python
nums.clear()
```

```python
nums
```

#### E). del() : Delete an items by index value.

```python
nums = [10,20,30,40]
print(nums)

del(nums[2])  # 30-position is 2
print(nums)
```

## Delete entire list

---

```python
del(nums)
```

```python
print(nums)
```

## 8. List Methods

----

|Method| Work|
|---|---|
|append(x)|Add item at end|
|extend(iterate)|Add multiple items|
|insert(i,x)|Insert at index position|
|remove(x)|Remove first occurrence|
|pop(i)|Remove item at index|
|pop()|Remove last item|
|clear()|Remove all items|
|del()|delete item by index|
|index(x)|Return index of value|
|count(x)|Count occurrence|
|sort()| Sort list|
|reverse()|Reverse list|
|copy()|Return shallow copy|

## 9. index() and count()

---


```python
nums = [10,20,30,40,20]
print(nums)
```

```python
print(nums.index(20))  # first occurrence value returned
```

```python
print(nums.count(20))
print(nums.count(10))
```

## 12. Copying List

---

A). Assignment Copy(Reference Copy)



```python
a = [10,20,30,40]
print(a)
```

```python
# Deep Copy
b = a
print(a)
```

```python
print(b)
```

```python
a.append(-50)
print(a)
```

```python
print(b)
```

### Shallow copy()

copy()
--

```python
a = [10,20,30,40]
print(a)


c = a.copy()
print(c)
```

```python
c.append(-500)
print(c)
```

```python
print(a)
```
