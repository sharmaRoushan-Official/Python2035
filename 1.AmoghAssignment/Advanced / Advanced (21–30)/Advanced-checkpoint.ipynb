21. Write a Python program to calculate *discount* based on purchase amount:

    * > 1000 → 20% discount
    * 500–1000 → 10% discount
    * <500 → No discount

```python
amount = int(input("Enter the purchase amount :"))
if amount > 1000:
    discount = 0.20
elif 500 <= amount <= 1000:
    discount = 0.10
else:
    discount = 0.0
discount_amount = amount * discount
final_price = amount - discount_amount

print("Discount amount:", discount_amount)
print("Final price after discount:", final_price)
```

25. Write a Python program to calculate the *electricity bill*:

    * ≤100 units → ₹1 per unit
    * 101–200 units → ₹1.5 per unit
    * > 200 units → ₹2 per unit

```python
units = int(input("Enter electricity units consumed: "))

if units <= 100:
    bill = units * 5
elif units <= 200:
    bill = (100 * 5) + ((units - 100) * 7)
else:
    bill = (100 * 5) + (100 * 7) + ((units - 200) * 10)

print("Electricity Bill = ₹", bill)

```

29. Write a Python program to classify *BMI (Body Mass Index)*:

    * <18.5 → Underweight
    * 18.5–24.9 → Normal
    * 25–29.9 → Overweight
    * ≥30 → Obese

```python
weight = float(input("Enter weight in kg: "))
height = float(input("Enter height in meters: "))

bmi = weight / (height**2)

print("Your BMI is:", round(bmi, 2))

if bmi < 18.5:
    print("Category: Underweight")
elif 18.5 <= bmi <= 24.9:
   print("Category: Normal")
elif 25 <= bmi <= 29.9:
    print("Category: Overweight")
else:
    print("Category: Obese")
```

30. Write a Python program to check whether an input string is *empty, **only spaces, or **contains characters*.

```python
def check_string(s):
    if s == "":
        return "Empty string"
    elif s.isspace():
        return "Only spaces"
    else:
        return "Contains characters"


user_input = input("Enter a string: ")

result = check_string(user_input)
print("Result:", result)
```
