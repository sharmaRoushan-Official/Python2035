Using For 

```python
for i in range(1,5):
    for j in range(i):
        print("*", end="")
    print()                       
```

```python
rows = 5
i = 1

while i <= rows:        
    j = 1
    while j <= i:       
        print("*", end="")
        j += 1
    print()             
    i += 1
```

```python
for i in range(5, 0, -1): 
    for j in range(i):     
        print("*", end="")
    print()                 
```

Using While 

```python
# Reverse Left Triangle using while loop

i = 5  # number of rows

while i >= 1:
    j = 1
    while j <= i:
        print("*", end="")
        j += 1
    print()
    i -= 1
```

4. Right Trianlge
    *
   **
  ***
 ****
*****

Using For

```python
n = 5

for i in range(1, n + 1):
    for j in range(n - i):
        print(" ", end="")
    for k in range(i):
        print("*", end="")
    print()
```

Using While 

```python
n = 5
i = 1

while i <= n:
    j = 1
    while j <= (n - i):
        print(" ", end="")
        j += 1

    k = 1
    while k <= i:
        print("*", end="")
        k += 1

    print()
    i += 1
```

5. Reverse Right Traiangle
*****
 ****
  ***
   **
    *

Using For

```python
n = 5

for i in range(n, 0, -1):          # controls rows
    for j in range(n - i):         # prints spaces
        print(" ", end="")
    for k in range(i):             # prints stars
        print("*", end="")
    print()                         # move to next line
```

Using While

```python
n = 5
i = n

while i >= 1:              # rows
    j = 0
    while j < (n - i):     # spaces
        print(" ", end="")
        j += 1

    k = 0
    while k < i:           # stars
        print("*", end="")
        k += 1

    print()
    i -= 1
```

Using For

10. Alphabet Triangle
A
AB
ABC
ABCD
ABCDE

Using For

```python
n = 5
alphabets = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

for i in range(1, n + 1):
    for j in range(i):
        print(alphabets[j], end="")
    print()
```

Using While

```python
n = 5
alphabets = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

i = 1

while i <= n:              # outer loop for rows
    j = 0
    while j < i:           # inner loop for characters
        print(alphabets[j], end="")
        j += 1
    print()                # next line
    i += 1
```

11. Repeated Alphabet
A

CCC
DDDD
EEEEE

Using For

```python
n = 5
alphabets = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

for i in range(n):
    for j in range(i + 1):
        print(alphabets[i], end="")
    print()
```

Using While

```python
n = 5
alphabets = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

i = 0

while i < n:              # outer loop for rows
    j = 0
    while j <= i:         # inner loop for repetition
        print(alphabets[i], end="")
        j += 1
    print()               # move to next line
    i += 1
```
