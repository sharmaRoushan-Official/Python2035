11. Print a countdown from 10 to 1

Using for 

```python
for i in reversed(range(1, 11)):
    print(i)
```

Using While 

```python
i = 10

while i >= 1:
    print(i)
    i = i - 1
```

12. Take numbers as input until the user enters 0, then print their sum.

Using For 

```python
total = 0

for i in range(100):
    num = int(input("Enter a number (0 to stop): "))
    if num == 0:
        break
    total += num

print("Sum =", total)
```

Using While

```python
total = 0

num = int(input("Enter a number (0 to stop): "))

while num != 0:
    total += num
    num = int(input("Enter a number (0 to stop): "))

print("Sum =", total)
```

13. Keep asking for a password until the correct one is entered

Using For

```python
password = "python123"

for i in range(100):  
    user_input = input("Enter password: ")
    
    if user_input == password:
        print("Access granted")
        break
```

Using While

```python
password = "python123"

user_input = input("Enter password: ")

while user_input != password:
    user_input = input("Wrong password. Try again: ")

print("Access granted")
```

14. Find the sum of digits of a number using a while loop.

Using For

```python
num = input("Enter a number: ")
total = 0

for i in range(len(num)):
    total += int(num[i])

print("Sum of digits =", total)
```

Using While

```python
num = int(input("Enter a number: "))
total = 0

while num > 0:
    digit = num % 10
    total += digit
    num = num // 10

print("Sum of digits =", total)
```

15. Reverse a string using a while loop.

Using For

```python
s = input("Enter a string: ")
rev = ""

for i in range(len(s)):
    rev = s[i] + rev

print("Reversed string =", rev)
```

Using While 

```python
s = input("Enter a string: ")
rev = ""

i = 0

while i < len(s):
    rev = s[i] + rev
    i += 1

print("Reversed string =", rev)
```
