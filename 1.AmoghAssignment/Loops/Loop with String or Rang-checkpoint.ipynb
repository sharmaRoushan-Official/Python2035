26. Count the number of vowels in a string.

Using For 

```python
text = "Hello World"
vowels = "aeiouAEIOU"

count = 0

for i in range(len(text)):
    if text[i] in vowels:
        count += 1

print("Number of vowels:", count)
```

Using While 

```python
text = "Hello World"
vowels = "aeiouAEIOU"

count = 0
i = 0

while i < len(text):
    if text[i] in vowels:
        count += 1
    i += 1

print("Number of vowels:", count)
```

27.Check if a string is a palindrome using a loop.

Using For

```python
text = "madam"
rev = ""

for i in range(len(text)):
    rev = text[i] + rev

if text == rev:
    print("Palindrome")
else:
    print("Not a palindrome")
```

Using While

```python
text = "madam"

i = 0
rev = ""

while i < len(text):
    rev = text[i] + rev
    i += 1

if text == rev:
    print("Palindrome")
else:
    print("Not a palindrome")
```

28. Print Fibonacci series up to n terms using a loop.

Using For

```python
n = 10

a = 0
b = 1

for i in range(n):
    print(a, end=" ")
    c = a + b
    a = b
    b = c
```

Using While 

```python
n = 10

a = 0
b = 1
i = 0

while i < n:
    print(a, end=" ")
    c = a + b
    a = b
    b = c
    i += 1
```

29. Find the GCD of two numbers using a loop.

Using For

```python
a = 24
b = 36
gcd = 1

for i in range(1, min(a, b) + 1):
    if a % i == 0 and b % i == 0:
        gcd = i

print("GCD is:", gcd)
```

Using While 

```python
a = 24
b = 36

i = 1
gcd = 1

while i <= min(a, b):
    if a % i == 0 and b % i == 0:
        gcd = i
    i += 1

print("GCD is:", gcd)
```
