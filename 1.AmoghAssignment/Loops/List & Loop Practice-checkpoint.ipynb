21. Find the maximum number in a list using a loop.

Using For 

```python
numbers = [10, 45, 3, 99, 27, 56]

max_num = numbers[0]

for num in numbers:
    if num > max_num:
        max_num = num

print("Maximum number is:", max_num)
```

Using While 

```python
numbers = [10, 45, 3, 101, 27, 56]

max_num = numbers[0]
i = 1

while i < len(numbers):
    if numbers[i] > max_num:
        max_num = numbers[i]
    i += 1

print("Maximum number is:", max_num)
```

22. Calculate the average of numbers in a list.

Using For

```python
numbers = [10, 20, 30, 40, 50]

total = 0
count = 0

for num in numbers:
    total += num
    count += 1

average = total / count

print("Average:", average)
```

Using While 

```python
numbers = [10, 20, 30, 40, 60]

total = 0
count = 0
i = 0

while i < len(numbers):
    total += numbers[i]
    count += 1
    i += 1

average = total / count

print("Average:", average)
```

23. Count how many even numbers are in a list.

Using For

```python
numbers = [1, 2, 3, 4, 5, 6, 7, 8, 10]

count = 0

for num in numbers:
    if num % 2 == 0:
        count += 1

print("Count of even numbers:", count)
```

Using While

```python
numbers = [1, 2, 3, 4, 5, 6, 7, 8, 10]

count = 0
i = 0

while i < len(numbers):
    if numbers[i] % 2 == 0:
        count += 1
    i += 1

print("Count of even numbers:", count)
```

24. Remove all negative numbers from a list using a loop.

Using For 

```python
numbers = [10, -5, 3, -1, 0, 7, -8, 12]

positive_numbers = []

for num in numbers:
    if num >= 0:
        positive_numbers.append(num)

print("List without negative numbers:", positive_numbers)
```

Using While 

```python
numbers = [10, -5, 3, -1, 0, 7, -8, 12]

i = 0

while i < len(numbers):
    if numbers[i] < 0:
        numbers.pop(i)  # remove negative number
    else:
        i += 1

print("List without negative numbers:", numbers)
```

25. Create a new list with squares of numbers from another list.

Using For

```python
numbers = [1, 2, 3, 4, 5]

squares = []

for num in numbers:
    squares.append(num * num)

print("Squares of numbers:", squares)
```

Using While 

```python
numbers = [1, 2, 3, 4, 5]

squares = []
i = 0

while i < len(numbers):
    squares.append(numbers[i] * numbers[i])
    i += 1

print("Squares of numbers:", squares)
```
