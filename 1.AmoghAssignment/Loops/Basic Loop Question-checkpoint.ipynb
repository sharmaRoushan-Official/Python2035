1. Print numbers from 1 to 10 using a for loop.

```python
for i in range(1, 11):
    print(i)
```

Using While Loop

```python
i = 1
j = 11

while i < j:
    print(i)
    i += 1
```

3. Print the table of a given number (e.g., input 5 → 5, 10, 15, … 50)

```python
num = int(input("Enter a number: "))

for i in range(1, 11):
    print(f"{num} x {i} = {num * i}")
```

Using While Loop

```python
num = int(input("Enter a number: "))

i = 1
j = 10

while i <= j:
    print(num * i)
    i = i + 1
```

4. Calculate the sum of numbers from 1 to n

```python
n = int(input("Enter a number: "))

sum = 0

for i in range(1, n + 1):
    sum = sum + i

print("Sum =", sum)
```

Using While 

```python
n = int(input("Enter n: "))

i = 1
sum = 0

while i <= n:
    sum = sum + i
    i = i + 1

print("Sum =", sum)
```

5. Calculate the factorial of a number using a loop.

```python
n = int(input("Enter a number: "))

fact = 1

for i in range(1, n + 1):
    fact = fact * i

print("Factorial =", fact)
```

Using While 

```python
n = int(input("Enter a number: "))

fact = 1
i = 1

while i <= n:
    fact = fact * i
    i = i + 1

print("Factorial =", fact)
```

7. Print the reverse of a given number (e.g., 123 → 321).

```python
n = int(input("Enter a number: "))

rev = 0
length = len(str(n))

for i in range(length):
    digit = n % 10
    rev = rev * 10 + digit
    n = n // 10

print("Reverse =", rev)
```

Using While 

```python
num = int(input("Enter a number: "))

reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

print("Reversed number:", reverse)
```

8. Count digits in a number using a loop.

```python
num = int(input("Enter a number: "))

num = abs(num)
count = 0

for i in range(num):
    num = num // 10
    count += 1
    if num == 0:
        break

print("Number of digits:", count)
```

Using While 

```python
num = int(input("Enter a number: "))

num = abs(num)
count = 0

while num > 0:
    num = num // 10
    count += 1

print("Number of digits:", count)
```

9. Print all characters of a string using a loop.

```python
str = input("Enter a string ")
for i in range(len(str)):
    print(str[i])
```

Using While 

```python
text = input("Enter a string: ")

i = 0

while i < len(text):
    print(text[i])
    i += 1
```

10. Print only vowels from a string input by the user.

```python
text = input("Enter a string: ")

vowels = "aeiouAEIOU"

for i in range(len(text)):
    if text[i] in vowels:
        print(text[i])
```
