# 1. Accepts any city and display their Monuments.

# cityName-------------------> Monument

# "Delhi"--------------------> ['Redfort','QutubMinar','LotusTemple']
# Agra-----------------------> "TajMahal"
# Jaipur---------------------> "HawaMahal/JalMahal"
# Amritsar-------------------> "GoldenTemple,Jalliawalabagh"
# Madurai--------------------> "Meenakshi Temple"
# Mysore---------------------> "Mysore palace"
# Mumbai --------------------> "Gateway of India"
# Sanchi---------------------> "Sanchi Stupa"
# Hydrabad-------------------> "Charminar"
# Gujarat -------------------> "Statue of Unity"

```python
def get_monuments(city):
    monuments = {
        "delhi": ["Red Fort", "Qutub Minar", "Lotus Temple"],
        "agra": ["Taj Mahal"],
        "jaipur": ["Hawa Mahal", "Jal Mahal"],
        "amritsar": ["Golden Temple", "Jallianwala Bagh"],
        "madurai": ["Meenakshi Temple"],
        "mysore": ["Mysore Palace"],
        "mumbai": ["Gateway of India"],
        "sanchi": ["Sanchi Stupa"],
        "hyderabad": ["Charminar"],
        "gujarat": ["Statue of Unity"]
    }

    city = city.strip().lower()

    return monuments.get(city, "No monuments found for this city")



print(get_monuments("Delhi"))
print(get_monuments("Agra"))
print(get_monuments("Hyderabad"))
```

# 2. A shop wil give you discount of 20%, if the cost of purchased is more than 2000, and if 
# purchased amount will be 1000 more than then it will discount 10%, And if purchased amout 
# is less than 1000- "No discount" as well print how much need more shopping 
# amount will to get minimum discount

```python
amount = int(input("Enter purchase amount: "))

if amount > 2000:
    discount = amount * 0.20
    print("20% discount applied")

elif amount > 1000:
    discount = amount * 0.10
    print("10% discount applied")

else:
    discount = 0
    print("No discount")
    print("You need", 1000 - amount, "more to get minimum discount")

final_amount = amount - discount
print("Final amount:", final_amount)
```

# 3.A compay will decided to give bonus to the employee according to the following criteria:
# Time period of the service -------------------> Bonus
# More then 10 Years ---------------------------> 10%
# >=6 and <=10 ---------------------------------> 8%
# <6 -------------------------------------------> 5%
# Ask the user for their salary and yers of service and then calcualte the bonus amount is, 
# and display the total salary after increment of bonus.

```python
salary = int(input("Enter salary: "))
years = int(input("Enter years of service: "))

if years > 10:
    bonus = salary * 0.10
elif years >= 6:
    bonus = salary * 0.08
else:
    bonus = salary * 0.05

total_salary = salary + bonus

print("Bonus:", bonus)
print("Total salary after bonus:", total_salary)
```

# 7. Accept the following from the user and calculate the percentage of the class attendence.
# Total number of working days.
# Total number of absent.
# if percentage is less than 75%, then student will not sit the exam.

```python
total_days = int(input("Enter total working days: "))
absent = int(input("Enter number of absent days: "))

present = total_days - absent
percentage = (present / total_days) * 100

print("Attendance percentage:", percentage)

if percentage < 75:
    print("You are NOT allowed to sit in the exam")
else:
    print("You are allowed to sit in the exam")
```

# 8. Calculate letters grades (A,B,C,D,E) based on score ranges (eg.-90-100:A,80-89:B,70-79:C,
# 60-69:D,50-59:E)

```python
marks = int(input("Enter your marks: "))

if marks >= 90:
    grade = "A"
elif marks >= 80:
    grade = "B"
elif marks >= 70:
    grade = "C"
elif marks >= 60:
    grade = "D"
elif marks >= 50:
    grade = "E"
else:
    grade = "Fail"

print("Grade:", grade)
```

# 15. Crete a simple calulator that performs basic arithmetic operations based on two user input.
# (^,*,/,-,+)

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
op = input("Enter operator (+, -, *, /, ^): ")

if op == "+":
    print(a + b)
elif op == "-":
    print(a - b)
elif op == "*":
    print(a * b)
elif op == "/":
    print(a / b)
elif op == "^":
    print(a ** b)
else:
    print("Invalid operator")
```
