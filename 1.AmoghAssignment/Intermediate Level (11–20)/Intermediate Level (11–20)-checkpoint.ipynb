13. Write a Python program to check if a number is *a single-digit, two-digit, or more than two digits*.

```python
num = int(input("Enter a number: "))
digits = len(str(abs(num)))
if digits == 1:
    print("It is a single-digit number.")
elif digits == 2:
    print("It is a two-digit number.")
else:
    print("It is more than two digits.")
```

15. Write a Python program to check whether a character is a *digit, **uppercase letter, **lowercase letter, or **special character*.

```python
ch = input("Enter a single character: ")
if len(ch) !=1:
    print("Please enter exactly one character.")
else:
    if ch.isdigit():
        print("It is a Digit")
    elif ch.isupper():
        print("It is uppercase letter")
    elif ch.islower():
        print("It Is lowercase letter")
    else:
        print(" It Is special letter")
```

20.Write a Python program to check if the entered number is *a multiple of 2, 3, both, or none*.

```python
num = int(input("Enter a number: "))

if num % 2 == 0 and num % 3 == 0:
    print("The number is a multiple of both 2 and 3.")
elif num % 2 == 0:
    print("The number is a multiple of 2 only.")
elif num % 3 == 0:
    print("The number is a multiple of 3 only.")
else:
    print("The number is not a multiple of 2 or 3.")
```
