### 7.Conditions in Python

----------

### 1. What is a conditional Statement?

A `conditional statement` allows a program to make decisions based on conditions(True/False)

> In Python, We use:
 * if
 * elif
 * else

### 2. Basic if Statement

Syntax:

if condition:
    
    # Code executes if condition is True.

```python
age = 18

if age>18:  # age=18 . 18>18
    print("You are elegible to vote.")  # True Block
```

```python
age = 18

if age>18:  # age=18 . 18>18
    print("You are elegible to vote.")  # True Block
else:  # runs directly when condition belongs False
    print("You are not elegible to vote")  # False block
```

```python
age = 18

if age>=18:  # age=18 . 18>18
    print("You are elegible to vote.")  # True Block
else:  # runs directly when condition belongs False
    print("You are not elegible to vote")  # False block
```

### Key Points:

* conditions must return True/False
* Indentation is mandatory
* No parentheses required(unlike C/java)

```python
age = 18

if (age>=18):  # age=18 . 18>18  # () is not required.
    print("You are elegible to vote.")  # True Block
else:  # runs directly when condition belongs False
    print("You are not elegible to vote")  # False block
```

#### 3. if-elif-else (ladder if)

syntax:

if condition1:
    
    # Block 1

elif condition2:
    
    # Block 2

elif condition3:
    
    # Block 3

elif condition4:
    
    # Block 4

else:
    
    ## default block

```python
marks = 75

if marks>= 90:  # False
    print("Grade A")
elif marks>=70:  # True
    print("Grade B")
elif marks>=50:
    print("Grade C")
else:
    print("Fail")
```

```python
marks = 45

if marks>= 90:  # False
    print("Grade A")
elif marks>=70:  # True
    print("Grade B")
elif marks>=50:
    print("Grade C")
else:
    print("Fail")
```

```python
marks = 101

if marks>= 90:  # False
    print("Grade A")
elif marks>=70:  # True
    print("Grade B")
elif marks>=50:
    print("Grade C")
else:
    print("Fail")
```

```python
marks = 66

if marks>= 90:  # False
    print("Grade A")
elif marks>=70:  # True
    print("Grade B")
elif marks>=50:
    print("Grade C")
else:
    print("Fail")
```

### Key Points:

* Multiple conditions can be checked
* Only first True condition exeuctes
* else is optional if all conditions goes False then it will excuted by default.

### 5. Nested if Statements

```python
age = 18
citizen = False

if age>=18:  # fist check the condition
    if citizen:  # True Block condition
        print("You are elegible to vote")
else:
    print("You are not elegible to vote")
```

```python
age = 18
citizen = True

if age>=18:  # fist check the condition
    if citizen:  # True Block condition
        print("You are elegible to vote")
    else:
        print("You are not elegible because you are not citizen of country.")
else:
    print("You are not elegible to vote")
```

### Key Points:

* if inside another if
* useful for multi-level decisions

### 7. Ternary Operator(Inline if-else)

syntax:

value_if_true if conditions else value_if_false

```python
age = 19

"Adult" if age>=18 else "Minor"
```

AND(and)

```python
age = 20

if age>=18 and age<=60:  # True,  True
    print("Working age")
else:
    print("Not Working age")
```

OR(or)


```python
day = "Sunday"

if day == "Saturday" or day == "Sunday":
    print("Weekend")
else:
    print("Weekday")
```

NOT(not)


```python
is_logged_in = False


if not is_logged_in:
    print("Please Login")
else:
    print("Not Requred to Login Already Login")
```

```python
day = 10

match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case 4:
        print("Thursday")
    case 5:
        print("Friday")
    case 6:
        print("Saturday")
    case 7:
        print("Sunday")
    case _:
        print("Invalid Day")
```
