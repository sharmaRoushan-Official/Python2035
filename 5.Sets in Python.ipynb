### 5. Sets in Python

1. Introduction to Sets

A `set` in Python is a collection of data type used to store multiple unique elements.

It is based on the mathematical concept of set.

---

### Key Characterstics

1. Unordred
2. Unidexed
3. Mutable(unhashable/can add/remove items)
4. Does not allow duplicate values
5. Elements must be immutable(Can't be change after creation/hashable)


Example

```python
s = {10,10,15,3,5,7}
print(s)
print(type(s))
```

#### Order is not gurateed.


### 2. Creating Sets

Method 1: Using Curly Barces {}

```python
s = {1,2,3,4}
print(s)
print(type(s))
```

Method 2: Using set() Constructor

```python
s = set([1,2,3,4,5,6])
print(s)
print(type(s))
```

Empty set cannot be created using {}



```python
s = {}
print(s)
print(type(s))
```

### 5. Accessing Set Elements 

Since Sets are unordered and Unidexed, we cannot access elements using index.



```python
s = {10,20,30}
print(s[0])
```

Error Occurs

Correct Way: Loop through the set

```python
s = {10,20,30}

for i in s:
    print(i)
```

### 10. Sets Operations

Python sets support mathematical operations.

1. Union
2. Intersection
3. Difference
4. Symmetric Different

|Method|Symbol|
|----|----|
|.union()|pipe - &#124; |
|.intersection()| Ampersand- & |
|.difference()| Subtract: -|
|.symmetric_difference()| Caret- ^|

```python
# Union
A = {1,2,3}
B = {3,4,5}

print(A)
print(B)
```

```python
# union

# {1,2,3,4,5}

# method()
print(A.union(B))
print(B.union(A))
```

```python
### Symbol- |
print(A | B)
print(B | A)
```

```python
## 2. Intersection - It return duplicate value

A = {1,2,3}
B = {3,4,5}

print(A)
print(B)

```

```python
# method 

print(A.intersection(B))
print(B.intersection(A))
```

```python
# symbol
print(A & B)
print(B & A)
```

```python
### 3. difference

A = {1,2,3}
B = {3,4,5}

print(A)
print(B)
```

```python
# method 

print(A.difference(B))
print(B.difference(A))
```

```python
### Symbol

print(A-B)
print(B-A)
```

```python
### 4. Symmetic Difference

A = {1,2,3}
B = {3,4,5}

print(A)
print(B)
```

```python
# method 

print(A.symmetric_difference(B))
```

### 11. Set Comparison Operations

##### Subset()

##### check if set is subset

```python
A = {1,2}
B = {1,2,3}
print(A)
print(B)
```

```python
# subset

print(A.issubset(B))
print(B.issubset(A)) # 3 key is extra which is not found in set A
```

```python
## superset

print(A.issuperset(B))  ## False- A is subset of B
print(B.issuperset(A))  ## True - B has extra element or keys from A and B all keys cannot found in A
```

### disjoint

Check if no common elements



```python
A = {1,2}
B = {3,4}

print(A.isdisjoint(B))
```

### 16. Built-in Functions with Sets

-----

|Function|Description|
|----|---|
