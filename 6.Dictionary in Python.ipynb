### 6. Dictionary in Python


A `dictionary` in Python is a built-in data type used to store data in the form of key-value paris.

It is one of the most used data structures becasue it allows fast searching, updating, and retrieving values.


### 1. Definition of Dictionary

A dictionary is written using Curly braces {} and contains elements in the format:

Key:Value

A dictionary is a collection of key-value pairs.

* Each key is unique.
* Each key maps to a value.
* Dictionaries are mutable(changeable/unhashable)
* They are written using curly braces {}.

```python
student ={
    "name":"Avnish",
    "age":12,
    "course":"Python"
}
print(student)
print(type(student))
```

* name,age,course = keys
* Avnish, 12,Python = Values

### 2. Why we Use Dictionaries

Use Dictionaries when data has a relationship between key and value.

Example Uses:

### 3. Characteeristics of Dictionary

## 1. Key must be unique



```python
data = {
    "name":"Amogh",
    "name":"Avnish"
}
print(data)
```

Notes:

Last value overwrites the previous one.

### 2. Keys must be immutable

Allowed:

* string
* number
* tuple

Not allowed:

* list
* dictionary
* set

```python
data = {
    1:"Python",
    "name":"Avnish",
    (1,2):"Tuple key"
}
print(data)
```

```python
data = {
    1:"Python",
    "name":"Avnish",
    (1,2):"Tuple key",
    {1,2}:"Set keys"
}
print(data)
```

```python
data = {
    1:"Python",
    "name":"Avnish",
    (1,2):"Tuple key",
    [1,2]:"Set keys"
}
print(data)
```

```python
data = {
    1:"Python",
    "name":"Avnish",
    (1,2):"Tuple key",
    {'a':'apple'}:"Dictionary keys"
}
print(data)
```

### 11. Looping Through Dictionary
