#### 1. What is a String in Python?

A `String` in Python is a Sequence of Unicode characters enclosed in quotes.

```python
s1 = "Hello"
s2 = 'World'
s3 = """Multi-line String"""
s4 = '''Multi-line Comment'''
print(s1)
print(s2)
print(s3)
print(s4)
```

### 7. Important String Method 
----
Case Conversion

```python
s = "pYTHon wORld" 
print("Original Text:",s)

# upper()
upperS = s.upper()
print(upperS)


# lower()
lowerS = s.lower()
print(lowerS)

# title
titleS = s.title()
print(titleS)

# capitalize
capitalizeS = s.capitalize()
print(capitalizeS)

# swapcase
swapcaseS = s.swapcase()
print(swapcaseS)
```

### Searching & Finding 


```python
s = "pYTHon wORld" 
print("Original Text:",s)

print(s.find("p"))  # returns index or -1
print(s.index('p')) # error if not found
```

```python
print(s.find("Z"))  # returns index or -1
```

```python
print(s.index('Z')) # error if not found
```

####  Checking Method 

```python
s = "Hello World" 
d = "123456"
space1 = " "
lower1 = 'hello'
upper1 = "HELLO"
print("Original Text:",s)

print(s.isalpha())  # it checks only alphabet characters 
print(s.isdigit())
print(d.isdigit()) # it checks only digit characters
print(s.isspace())
print(space1.isspace()) # it checks only space value
print(s.islower())  # it checks only lower characters
print(lower1.islower())
print(s.isupper()) # it checks text is upper or not?
print(upper1.isupper())
```

## 9. Escape Character 
---
|Escape|Meaning|
|----|----|
|`\n`| New Line|
|`\t`| Tab|

|`\'`| Single Quote|
