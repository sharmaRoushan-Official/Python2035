## Rules for creating Variable in Python

1. Variable Name must start with 

✅ Letter (a-z,A-Z)

✅ Underscore(_)

```python
name = "Raj"
_age = 25
city1 = "Vasant Vihar"
print(name)
print(_age)
print(city1)
```

2. Variable Name cannot start with Number


```python
1name = "Raj"
```

3. No Spaces Allowed

```python
first Name = "Raj"
```

4. Special Symbols are Not allowed


```python
price@ = 100
```

5. Python keyword cannot be used as variable

```python
class = 10
```

Beacuse class is `reserved` keyword

```python
is = 100
```

6. Varibale Name are Case Sensitive

```python
name = "A"
Name = 'B'

print(name)
print(Name)
```
