### Operators in Python

1. Assignment Operator: [=]

2. Arithmetic Operator: [+,-.*,/,//,%,**]
    1. Uniary Operator: [++,--]
       
            > It works on single value
       
            > Pre and Post
       
            > Pre Increment: ++i   [one plus in any value]
       
            > Post Increment: i++  [One plus in any value]
       
            > Pre Decrement: --i   [Minus one value in variable]
       
            > Post Decrement: i--  [Minus one value in Variable]
       
    2. Binary Operator: [+,-,*,/,//,%,**]
  
       
            > It works at least two operands.
  
       
            > Ex: 10+20+50-30 = 30, 10-30= -20
  
       
3. Comparison/Relation Operator: [<,>,<=,>=,==,!=]

4. Logical Operator

        > Basic Gates: [AND, OR, NOT]

        > Universal Gates: [NAND, NOR]

        > Exclusive OR Gate: [XOR]

        > Exclusive NOR Gates: [XNOR]

6. Identity Operator: [is or is not]

7. Membership Operator: [in or not in]

### Data TypeCasting
---


```python
# to convert number: integer value to float and string

num = 10   # num is variable and = is assignment operator 10 is integer value
print(num)
print(type(num))

# float
float1 = float(num) 
print(float1)
print(type(float1))


# string

str1 = str(num)
print(str1)
print(type(str1))
```

```python
## number:float 125.456 convert this into string, integer

# original value
num2 = 125.456
print(num2)
print(type(num2))


# integer
int1 = int(num2)
print(int1)
print(type(int1))

# string 
str1 = str(num2)
print(str1)
print(type(str1))
```
